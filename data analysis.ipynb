```python
acc = {}
pre = {}
rec = {}
f1 = {}
```

```python
classes = []
def readfile(file):
    txt = f"{(file)}.csv"
    df = pd.read_csv(txt)
    df = np.array(df)
    
    a_name = f"{(file)}_acc" 
    p_name = f"{(file)}_pre"
    r_name = f"{(file)}_rec"
    f1_name = f"{(file)}_f1"
    
    acc[a_name] = np.round(df[2][1],3)
    pre[p_name] = np.round(df[1][1],3)
    rec[r_name] = np.round(df[1][2],3)
    f1[f1_name] = np.round(df[1][3],3)
    
```

```python
readfile("n128")
readfile("n512_512")
readfile("n256_512")
readfile("n512_64_512")
readfile("n256_64_16_64")
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
        
# creating the dataset
Neurons = list(acc.keys())
acc = list(acc.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(Neurons, acc, color ='black', 
        width = 0.4)
addlabels(Neurons, acc) 
plt.xlabel("Number of Neurons")
plt.ylabel("Accuracy")
plt.title("The Accurracy with varying neurons & hidden layers")
plt.show()
```
